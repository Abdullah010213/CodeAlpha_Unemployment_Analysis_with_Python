```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
data.sample(10)
```

```python
print("Duplicate Rows: ", data.duplicated().sum())
```

```python
print("Missing Values:")
data.isna().sum()
```

```python
print("Display all Columns: ")
data.columns
```

```python
data = data.drop(columns=['Region.1', ' Frequency'])
print("Droped Region.1 and Frequency cloumns")
```

```python
print("Data set Info: ")
data.info()
```

```python
data.sample(10)
```

```python
print("Summary Statistics:")
data.describe()
```

```python
# Overall Unemployment Trend Analysis
plt.figure(figsize=(14, 7))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=data, color='red')
plt.title('Monthly Unemployment Rate Trend in India (2020)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Regional Analysis
# Top 5 states with highest unemployment
top_states = data.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().nlargest(5)
plt.figure(figsize=(10, 6))
top_states.plot(kind='bar', color='purple')
plt.title('Top 5 States with Highest Average Unemployment Rate')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
data[' Date'] = pd.to_datetime(data[' Date'])
```

```python
data[' Date'].isna().sum()
```

```python
data['Month'] = data[' Date'].dt.month_name()
data['Year'] = data[' Date'].dt.year
```

```python
data.sample(5)
```

```python
# COVID timeline comparison (pre-lockdown vs post-lockdown)
covid_impact = data.groupby('Month')[[' Estimated Unemployment Rate (%)', ' Estimated Labour Participation Rate (%)']].mean()
covid_impact = covid_impact.reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September'])

plt.figure(figsize=(14, 7))
sns.lineplot(data=covid_impact, markers=True)
plt.title('Monthly Employment Trends During 2020')
plt.xlabel('Month')
plt.ylabel('Percentage (%)')
plt.grid(True)
plt.show()
```

```python
# Correlation Analysis
corr_matrix = data[[' Estimated Unemployment Rate (%)',
                ' Estimated Employed',
                ' Estimated Labour Participation Rate (%)']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Employment Factors')
plt.show()
```

```python
# Employment vs Unemployment Relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x=' Estimated Labour Participation Rate (%)',
                y=' Estimated Unemployment Rate (%)',
                data=data,
                hue='Region',
                palette='viridis')
plt.title('Labour Participation vs Unemployment Rate')
plt.show()
```

```python
# State-wise Detailed Analysis (Maharashtra)
maharashtra_data = data[data['Region'] == 'Maharashtra']

plt.figure(figsize=(12, 6))
plt.plot(maharashtra_data[' Date'],
         maharashtra_data[' Estimated Unemployment Rate (%)'],
         marker='o')
plt.title('Maharashtra Unemployment Rate Timeline')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Prepare data
X = data[['longitude', 'latitude', ' Estimated Labour Participation Rate (%)']]
y = data[' Estimated Unemployment Rate (%)']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)

# Train model
model = RandomForestRegressor(random_state=2)
model.fit(X_train, y_train)

# Evaluate
print("Model Score:", model.score(X_test, y_test))
```
